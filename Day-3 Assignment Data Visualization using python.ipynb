## Question 1


## Plot using sns.relplot, line plot, box plot on fmri data set

```python
fmri = sns.load_dataset('fmri')
fmri.head(5)
```

```python
fmri.describe()
```

```python
sns.relplot(x = 'timepoint', y = 'signal', hue = 'region', style = 'event', size = 'subject',palette ="ch:rot=-.50,hue=3,light=.45",data = fmri ,height = 7)
```

# Line Plot

```python
sns.relplot(x="timepoint",y="signal",kind="line",hue ="region",style= "event",palette ="ch:rot=-.50,hue=3,light=.45",data=fmri,height = 7)
```

```python
sns.relplot(data=fmri, x="timepoint", y="signal", col="region",hue="event", style="event", kind="line")
```

# Box Plot

```python
sns.set(style='whitegrid')
sns.boxplot(x="timepoint",y="signal",data=fmri)
```

```python
sns.boxplot(x="timepoint",y="signal",hue = "region",data=fmri)
```
