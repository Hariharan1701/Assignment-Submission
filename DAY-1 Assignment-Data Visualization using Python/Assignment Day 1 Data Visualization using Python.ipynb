```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

## Question 1


## plot a line plot between a and b:

```python
a = np.arange(40,50)
b = np.arange(50,60)

plt.plot(a , b,'co--',linewidth= 2,markersize= 5)
plt.title('Line plot between a and b' , color='Red')
plt.xlabel('a')
plt.ylabel('b')
plt.show()
```

# Question 2

## Plot a line plot showing the sales trend in company 1 and 2:

```python
days = [1,2,3,4,5,6,7] #days of d week
sales_1 = [160,150,140,145,175,165,180] #sales of company1
sales_2 = [70,90,160,150,140,145,175]  #sales of company2

plt.title("Line plot showing the sales trends of company 1 & 2",color='blue')
plt.plot(days, sales_1,'g*-', label = 'Sales of company1')
plt.plot(days, sales_2,'r*-', label = 'Sales of company2')
plt.xlabel("Days of the week")
plt.ylabel("Company sales")
plt.legend(loc='lower right')
plt.show()
```
